```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 定义函数
def f(x):
    Q = np.array([[1, 0], [0, 10]])
    return 0.5 * x.T @ Q @ x + 10

# 返回f(x)梯度值
def grad_f(x):
    Q = np.array([[1, 0], [0, 10]])
    return Q @ x

def plt_f(X, Y):
    x = np.array((X,Y))
    Q = np.array([[1, 0], [0, 10]])
    return 0.5 * x.T @ Q @ x + 10
```

```python
# 作图方法，f(x)变化率和f(x)三维图像
def show_plt(x_list, y_list, f_list, fl_list, S):
    plt.figure(figsize=(10,6))
    plt.plot(range(len(fl_list)),fl_list,'r')
    plt.title(S)
    plt.xlabel('Iterations')
    plt.ylabel('f(x) -f(x*)')
    # plt.show()

    fig = plt.figure()
    ax3d = fig.add_subplot(projection='3d')  #创建3d坐标系
    plt.rcParams['axes.unicode_minus'] = False

    Q = np.array([[1, 0], [0, 10]])
    x_range = np.linspace(-1,1, 50)
    y_range = np.linspace(-1,1,50)
    X,Y = np.meshgrid(x_range, y_range)
    Z = 0.5 * (X*Q[0][0]*X + Y*Q[1][1]*Y) + 10
    ax3d.plot_surface(X,Y,Z)
    ax3d.scatter(x_list,y_list,f_list, c='r', marker='o')
    ax3d.view_init(elev=32, azim=41)
    plt.show()
```

```python
# 精确线搜索法
def line_search(x):
    Q = np.array([[1, 0], [0, 10]])

    # 第k步步长
    return np.linalg.norm(Q@x, ord=2)**2 / ((Q@x).T @ Q @ (Q @ x))

def Acc_line_search(x0):
    xi = x0

    # 结束条件
    tol = 1e-10
    pre_tol = 1

    cur_y = f(xi)
    max_iter = 1000

    # 存储数据
    x_list = [x0[0]]
    y_list = [x0[1]]
    f_list = [f(x0)]
    fl_list = []

    for i in range(0, max_iter):
        gi = grad_f(xi)

        # 更新步长和x的位置
        ti = line_search(xi)
        xi = xi - ti * gi
        x_list.append(xi[0])
        y_list.append(xi[1])

        pre_y = cur_y
        cur_y = f(xi)
        f_list.append(cur_y)
        pre_tol = abs(pre_y -  cur_y)
        if i > 0:
            fl_list.append(pre_tol)
        if pre_tol < tol:
            break

    show_plt(x_list, y_list, f_list, fl_list, 'Accurate line search')
    print(xi.reshape([-1, 1]))
```

```python
# 固定步长法
def gradient_descent(x0):
    xi = x0
    L = 10

    # 结束条件
    tol = 1e-10
    pre_tol = 1
    
    cur_y = f(xi)
    max_iter = 1000

    # 存储数据
    x_list = [x0[0]]
    y_list = [x0[1]]
    f_list = [f(x0)]
    fl_list = []

    for i in range(0, max_iter):
        gi = grad_f(xi)

        # 更新x的位置
        ti = 1/L
        xi = xi - ti * gi
        x_list.append(xi[0])
        y_list.append(xi[1])

        pre_y = cur_y
        cur_y = f(xi)
        f_list.append(cur_y)
        pre_tol = abs(pre_y - cur_y)
        if i > 0:
            fl_list.append(pre_tol)
        if pre_tol < tol:
            break

    show_plt(x_list, y_list, f_list, fl_list, 'Gradient Descent')
    print(xi.reshape([-1, 1]))
```

```python
x0_1 = np.array([1.0, 1.0])
Acc_line_search(x0_1)
gradient_descent(x0_1)
```


1. 可以看出精确线搜索法可以在几步就靠近最优点,更快的满足条件;

2. 而固定步长梯度下降法是逐渐靠近最优点,由于步长固定，可以看出从第二步开始可以连成一条三维线，相比之下接近最优点更慢。

```python
# Goldstein准则非精确先搜索算法
# 通过二分法实现
def n_line_search(x, d):
    a = 0
    b = 1

    tol = 1e-6
    max_iter = 100

    for i in range(0, max_iter):
        c = (a+b) / 2
        fc = f(x + c * d)
        gc = grad_f(x + c * d)
        # 使用Armijo条件检查c是否是f和d的极小值
        if fc <= f(x) + c * tol * gc.T @ d:
            return c
        else:
            if gc.T @ d < 0:
                a = c
            else:
                b = c

    return c


def n_Acc_line_search(x0):
    xi = x0

    # 结束条件
    tol = 1e-10
    pre_tol = 1

    cur_y = f(xi)
    max_iter = 1000

    # 存储数据
    x_list = [x0[0]]
    y_list = [x0[1]]
    f_list = [f(x0)]
    fl_list = []

    for i in range(0, max_iter):
        gi = grad_f(xi)

        # 更新步长和x的位置
        ti = n_line_search(xi, -gi)
        # ti = line_search_wolfe2(f = f, myfprime=grad_f, xk = xi, pk = -gi)[0]
        if ti == None:
            break


        xi = xi - ti * gi
        x_list.append(xi[0])
        y_list.append(xi[1])

        pre_y = cur_y
        cur_y = f(xi)
        f_list.append(cur_y)
        pre_tol = abs(pre_y -  cur_y)
        if i > 0:
            fl_list.append(pre_tol)
        if pre_tol < tol:
            break

    show_plt(x_list, y_list, f_list, fl_list, 'Accurate line search')
    print(xi.reshape([-1, 1]))
```

```python
x0_1 = np.array([1.0, 1.0])
n_Acc_line_search(x0_1)
```

**相比精确线搜索法，非精确线搜索速度更慢，但是比固定步长法更快;**

**不过非精确线搜索没有像精确线搜索法那样基本接近最优点，它也是有一个渐进的过程，但是更加没有规律；**

**非精确线搜索法主要是解决精确线搜索可能找不到最优点的缺点，不过方法有很多。**
